```python
i = 0
wb_dataframe_list_csv_n = []
for dataframe in wb_dataframe_list_csv:
  temp_df = dataframe.copy()
  temp_df_n = temp_df.rename({'Value': wb_file_names[i] + ' Value'}, axis='columns')
  wb_dataframe_list_csv_n.append(temp_df_n)
  i = i+1
```

```python
i = 0
for new_dataframe in wb_dataframe_list_csv_n :
    for col in new_dataframe.columns:
      if col == 'Indicator Name' or col == 'Indicator Code' or col == 'Country Name' or col == 'Country Code' or col == 'Year' or col == 'Disaggregation':
        continue
      print(col)
```

Cleaning

```python
i = 0
for i, new_dataframe in enumerate(wb_dataframe_list_csv_n) :
  wb_dataframe_list_csv_n[i]=new_dataframe.drop(labels='Disaggregation', axis=1)
```
